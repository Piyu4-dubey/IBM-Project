```python
data=pd.read_csv('/content/sample_data/MentalHealth.csv')
```

```python
data=data[data.isnull().sum(axis=1)<=2]
```

```python
data
```

```python
#duplicacy check

if data.duplicated().any():
  data=data.drop_duplicates()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python
data.head(10)
```

```python
# Convert Yes/No to High/Low
data["Mental_Health_Risk"] = data["Mental_Health_Risk"].map({
    "Yes": "High",
    "No": "Low"
})
# Example logic: risk is moderate if stress is high but activity is also high
def assign_risk(row):
    if row["Workload_Stress_Score"] > 7 and row["Physical_Activity"] > 7:
        return "Moderate"
    elif row["Mental_Health_Risk"] == "High":
        return "High"
    else:
        return "Low"

data["Mental_Health_Risk"] = data.apply(assign_risk, axis=1)
```

```python
risk_mapping = {
    "Low": 0,
    "Moderate": 1,
    "High": 2
}

data["Mental_Health_Risk"] = data["Mental_Health_Risk"].map(risk_mapping)
```

```python
eating_map = {
    "Skipped": 0,
    "Irregular": 1,
    "Balanced": 2
}

data["Eating_Habit"] = data["Eating_Habit"].map(eating_map)
```

```python
import pandas as pd

data_df = pd.DataFrame(data)  # âœ…
data_df.to_csv("MentalHealth.csv", index=False)
```
